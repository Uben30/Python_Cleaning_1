```python
import pandas as pd
import matplotlib as mlp
import plotly as plt
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df=pd.read_html("https://fbref.com/en/comps/9/Premier-League-Stats")
df
```

```python
df_temp=df[1]
# get the first header row
header1 = df_temp.columns.get_level_values(0)

# get the second header row
header2 = df_temp.columns.get_level_values(1)

# remove the comma from the second header row
header2 = header2.str.replace(',', '')

# combine the two header rows and assign to the columns attribute
df_temp.columns = header1 + '_' + header2

df_temp
```

```python
df_temp=df_temp.rename(columns={'Unnamed: 0_level_0_Rk':'Ranking','Unnamed: 1_level_0_Squad':'Squad'})
```

```python
df_temp
```

```python
df_temp=df_temp.drop('index_',axis=1)
```

```python
df_temp
```

```python
df_temp.rename(columns={'Home_Home MP':'Home_MP'})
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_temp['Away_Pts/MP'], y=df_temp['Home_Pts/MP'], text=df_temp['Squad'], mode='markers+text', textposition='top center',
                         marker=dict(size=10, color='blue')))
fig.update_layout(title='Away & Home Points Per 90', xaxis_title='Away points per 90', yaxis_title='Home Points per 90')
fig.show()
```

```python
df_temp['H & A Goals Diff'] = df_temp['Home_GF'] - df_temp['Away_GF']
```

```python
df_temp
```

```python
# Sorting the df in descending order based of Home goals Scored, to get the bar chart in a descending order
df_temp_sorted=df_temp.sort_values('Home_GF',ascending=False)

# create a bar chart trace for home goals
trace_home_gf = go.Bar(x=df_temp_sorted['Squad'], y=df_temp_sorted['Home_GF'], name='Home GF',text=df_temp_sorted['Home_GF'],textposition='inside')

# create a bar chart trace for away goals
trace_away_gf = go.Bar(x=df_temp_sorted['Squad'], y=df_temp_sorted['Away_GF'], name='Away GF',text=df_temp_sorted['Away_GF'],textposition='outside')

# create a line chart trace for H & A goals diff, use yaxis parameter to enable a secondary y-axis using input y2
trace_ha_goals_diff = go.Scatter(x=df_temp_sorted['Squad'], y=df_temp_sorted['H & A Goals Diff'], name='H & A Goals Diff',
                                 yaxis='y2', line=dict(color='red'),text=df_temp_sorted['H & A Goals Diff'], textposition='top center')

# create the figure
fig = go.Figure(data=[trace_home_gf, trace_away_gf, trace_ha_goals_diff])

# update the layout, adding secondary y-axis, giving its position toward the right and overlaying attribute sets the second y-axis to overlay the first y-axis
fig.update_layout(title='Goals Scored and Goal Difference',
                  xaxis_title='Team', yaxis_title='Goals Scored',
                  yaxis2=dict(title='H & A Goals Diff', overlaying='y', side='right'))

# show the chart
fig.show()
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://fbref.com/en/comps/9/stats/Premier-League-Stats#stats_standard'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

tables = soup.find_all('table')  # find all table elements in the page

for table in tables:
    thead = table.find('thead')  # find the table header row
    headers = [th.text.strip() for th in thead.find_all('th')]  # extract the text from each header cell
    print(headers)

    tbody = table.find('tbody')  # find the table body
    rows = tbody.find_all('tr')  # find all rows in the table body

    for row in rows:
        cells = row.find_all('td')  # find all cells in the row
        values = [cell.text.strip() for cell in cells]  # extract the text from each cell
        print(values)


```
